```python
#filtro de comparación
F6=Dataf[Dataf["Periodo"] >2005]
F6
```

```python
#Filtro Y
F7=Dataf[(Dataf["Periodo"] >2005) & (Dataf['Sexo']=='Mujeres')]
F7
```

```python
#Filtro o
F8=Dataf[(Dataf["Periodo"] >2005) | (Dataf['Sexo']=='Mujeres')]
F8
```

```python
#Filtro o
F9=Dataf[~(Dataf["Periodo"] >2005) & ~ (Dataf['Sexo']=='Mujeres')]
F9
```
